```python
from matplotlib import pylab
```

```python
print(pylab.__version__)
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,25)
y = x * x + 2
print(x)
print(y)
```

```python
pylab.plot(x,y,'tomato')
```

```python
pylab.subplot(1,2,1) #(rows, cols, indexes)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,2)
pylab.plot(y,x,'g*-')
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

axes = fig.add_axes([0.5, 0.1, 0.8, 0.8])

axes.plot(x,y)
```

```python
#again we can draw subgraphs

fig,axes = plt.subplots(nrows = 1, ncols = 2)

for ax in axes:
    ax.plot(x,y)
```

```python
# draw one graph inside another graph

fig = plt.figure(dpi = 1000)

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y,'tomato')
axes2.plot(y,x)
```

```python
fig, axes = plt.subplots(dpi = 150)
axes.plot(x, x**2,alpha = 0.5,linewidth= 0.8)
axes.plot(x, x+2,linestyle = ':')
axes.plot(x, x+3,linestyle='-.')
axes.grid(True)
```

```python
fig, axes = plt.subplots(dpi = 150)
axes.plot(x, x**2,alpha = 0.5,linewidth= 0.8)
axes.plot(x, x+2,linestyle = ':')
axes.plot(x, x+3,linestyle='-.')
axes.grid(True)
axes.set_ylim([0,40])
axes.set_ylim([0,4])
```

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1,4, figsize = (16,5))

axes[0].set_title('scatter plot')
axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))

axes[1].set_title('step')
axes[1].step(n, n**2)

axes[2].set_title('bar')
axes[2].bar(n, n**2)

axes[3].set_title('fill between')
axes[3].fill_between(n, n**2,)
```

```python
#radar chart

fig = plt.figure(figsize = (6,6))

ax = fig.add_axes([0,0,.8,1], polar = True)

t = np.linspace(0, 2* np.pi, 100)
ax.plot(t,t, color = 'tomato')
```

```python
#histogram

n = np.random.randn(100000)
fig, ax = plt.subplots(1,2,figsize = (12,4))

ax[0].set_title('default histogram')
ax[0].hist(n)

ax[1].set_title('cumulative histogram')
ax[1].hist(n,bins = 5, cumulative = True, color = 'orange')
```

```python
#contour maps

import matplotlib.cm as cm # cm = color map

delta = 0.025
x = np.arange(-3,3,delta)
y = np.arange(-2,2,delta)
X,Y = np.meshgrid(x,y)
z1 = np.exp(-X**2 - Y**2)
z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)

z = (z1 - z2) * 2
```

```python
fig, ax = plt.subplots()
cs = ax.contour(X, Y, z)
ax.clabel(cs, inline = 2, fontsize = 10)
ax.set_title('contour map')
```

```python
#3D MAP

from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
fig = plt.figure(figsize = (14,6))

ax = fig.add_subplot(1,2,1, projection = '3d')
ax.plot_surface(X,Y,z, rstride = 4, cstride = 4, linewidth = 0)
```
