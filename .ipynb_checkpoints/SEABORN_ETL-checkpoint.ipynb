```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
# %matplotlib inline
# !pip install mpld3
# import mpld3
# mpld3.enable_notebook()
```

```python
df = pd.read_csv('Pokemon.csv',index_col = 0)
```

```python
df.head()
```

```python
ax = sns.lmplot(x = 'Attack',y = 'Defense', data = df)
sns.set_style('darkgrid')
```

```python
ax = sns.lmplot(x = 'Attack',y = 'Defense', data = df,fit_reg = False, hue = 'Stage')
```

```python
#Box Plot
sns.boxplot(data = df)
```

```python
stats_df = df.drop(['Total','Stage','Legendary'], axis = 1)

sns.boxplot(data = stats_df)
```

```python
sns.violinplot(x = 'Type 1', y = 'Attack', data = df)
```

```python
sns.violinplot(x = 'Type 2', y = 'Defense', data = df)
```

```python
import matplotlib.pyplot as plt
```

```python
new_stats = pd.DataFrame(list(zip(df.Attack,df.Defense)),columns = ['Attack','Defense'])
```

```python
new_stats.head()
```

```python
plt.scatter(new_stats.Attack, new_stats.Defense, edgecolors = ['red','blue'])
```

```python
sns.swarmplot(x = 'Type 1', y = 'Attack', data = df)
```

```python
plt.figure(figsize = (10,6))

sns.violinplot(x = 'Type 1', y = 'Attack', data = df, inner = None)

sns.swarmplot(x = 'Type 1', y = 'Attack', data = df, color = 'k', alpha = 0.7)

plt.title('Attack by Type')
```

```python
stats_df.head()
```

```python
melted_df = pd.melt(stats_df, id_vars = ['Name','Type 1','Type 2'], var_name = 'Stat')
melted_df.head()
```

```python
sns.swarmplot(x = 'Stat', y = 'value', data = melted_df, hue = 'Type 1',dodge = True)

plt.ylim(0,280)
plt.legend(bbox_to_anchor = (1,1), loc = 'best')
```

```python
#heatmap
corr = stats_df.corr()

sns.heatmap(corr)
```

```python
#histogram a.k.a Distribution plot

sns.distplot([df.Defense])
```

```python
sns.distplot([df.Attack])
```

```python
#count plot or Bar plot
sns.countplot(x = 'Type 1', data = df)

#rotate x-labels
plt.xticks(rotation = -45)
```

```python
#factor plot

g = sns.factorplot(x = 'Type 1', y = 'Attack', data = df, hue = 'Stage', col = 'Stage')

g.set_xticklabels(rotation = -45)
```

```python
#density plot

sns.kdeplot( df.Defense,df.Attack,)
```

```python
#joint dist. plot

sns.jointplot(x = 'Attack', y = 'Defense', data = df)
```
